```python
#libraries for data analysis
import pandas as pd
import numpy as np
#libraries for normality tests
from scipy.stats import shapiro
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
```

```python
#import the User Datasets
xls1 = pd.ExcelFile('Bounce_Rate_2018.xlsx')
xls2 = pd.ExcelFile('Bounce_Rate_2019.xlsx')
Bounce_Rate_2018 = pd.read_excel(xls1, 'Dataset1')
Bounce_Rate_2019 = pd.read_excel(xls2, 'Dataset1')
```

```python
#Remove the dates for ease
Bounce_Rate_2018=Bounce_Rate_2018['Bounce Rate']
Bounce_Rate_2019=Bounce_Rate_2019['Bounce Rate']
```

```python
#convert datasets into arrays
Bounce_Rate_2018=Bounce_Rate_2018.values
Bounce_Rate_2019=Bounce_Rate_2019.values
```

```python
#plot both datasets using qqplot to check for normality
qqplot(Bounce_Rate_2018, line='s')
pyplot.show()
```

```python
qqplot(Bounce_Rate_2019, line='s')
pyplot.show()
```

### Both datasets look normal since they fall on the qq plot

```python
# normality test with Shapiro
stat, p = shapiro(Bounce_Rate_2018)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

```python
stat, p = shapiro(Bounce_Rate_2019)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

### Since not both datasets are normaly distributed non - parametric tests should be used.

```python
#import ranksums (Wilcoxon-Mann-Whitney (WMW) rank sum test) to test wether the difference between the two datasets is significant
from scipy.stats import ranksums
```

```python
stat, p =  ranksums(Bounce_Rate_2018,Bounce_Rate_2019)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('The difference between the two datasets is not significant (fail to reject H0)')
else:
	print('The difference between the two datasets is significant (reject H0)')
```

## We conclude that there is significant difference between the two datasets
